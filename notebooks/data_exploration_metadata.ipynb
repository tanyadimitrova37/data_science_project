# Recruitment Data dataset

```python
# Reading the dataset
```

```python
df = pd.read_csv("data/recruitment_data.csv")
```

```python
df
```

```python
df.info()
```

It looks like the dataset has no missing values and all fields data types make sense. Some of the variables could be converted to categorical but I will leave them for now. 

```python
df.describe()
```

Distribution of key features looks reasonable (Gender, Age, Scores, Experience, etc.)

The column names are currently CamelCase. I will convert them to snake_case.

```python
df.columns = df.columns.str.strip().str.replace(r'(?<!^)(?=[A-Z])', '_', regex = True).str.lower()
```

I will drop unnecessary columns.

```python
df = df.drop(columns = "distance_from_company")
df = df.drop(columns = "previous_companies")
df = df.drop(columns = "recruitment_strategy")
```

I will convert the gender, eduaction_level, and hiring_decision columns to categorical variables and replace the digits with the corresponding values.

```python
df['gender'] = df['gender'].replace({0: 'Male', 1: 'Female'})
df['education_level'] = df['education_level'].replace({
    1: "Bachelor's",
    2: "Bachelor's",
    3: "Master's",
    4: "PhD"
})
df['hiring_decision'] = df['hiring_decision'].replace({0: 'Not Hired', 1: "Hired"})
```

```python
df['gender'] = df['gender'].astype('category')
df['education_level'] = df['education_level'].astype('category')
df['hiring_decision'] = df['hiring_decision'].astype('category')
```

```python
df.to_csv('data/cleaned_recruitment_data.csv', index = False)
```
